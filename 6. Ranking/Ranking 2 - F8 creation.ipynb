```python
import scipy
import numpy as np
import gensim
import re
import pandas as pd
from sklearn import feature_extraction
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = stopwords.words('english')
```

```python
lda_df = pd.read_pickle("df_cleaned.pkl")
```

```python
lda_df.head()
```

```python
paragraphs = lda_df['Paragraphs_cleaned'].tolist()
```

```python
tokenizer = RegexpTokenizer(r'\w+')
p_stemmer = PorterStemmer()
texts = []
```

```python
for i in paragraphs:
    
    # clean and tokenize document string
    raw = i.lower()
    tokens = tokenizer.tokenize(raw)

    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in stopwords]
    
    # stem tokens
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    
    # add tokens to list
    texts.append(stemmed_tokens)
```

```python
# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)
# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
# generate LDA model 100 topics
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=100, id2word = dictionary, passes=20)
```

```python
lda_corpus = [max(prob,key=lambda y:y[1])
                for prob in ldamodel[corpus] ]
playlists = [[] for i in range(100)]
for i, x in enumerate(lda_corpus):
    playlists[x[0]].append(paragraphs[i])
```

```python
docTopicProbMat = ldamodel.get_document_topics(corpus,minimum_probability=0)
listDocProb = list(docTopicProbMat)
```

```python
probMatrix = np.zeros(shape=(len(df["Paragraphs"]),100))
for i,x in enumerate(listDocProb):      #each document i
    for t in x:     #each topic j
        probMatrix[i, t[0]] = t[1] 
```

```python
matrix100 = pd.DataFrame(probMatrix)
```

```python
articles1 = list()
paragraphs1 = list()
articles2 = list()
paragraphs2 = list()
tfidf_cosine_a = list()
tfidf_bigram_cosine_a = list()
tfidf_trigram_cosine_a = list()
count_vec_cosine_a = list()
lda_50topics_cosine_a = list()
lda_100topics_cosine_a = list()
lda_200topics_cosine_a = list()
label = list()

articles1b = list()
paragraphs1b = list()
articles2b = list()
paragraphs2b = list()
tfidf_cosine_b = list()
tfidf_bigram_cosine_b = list()
tfidf_trigram_cosine_b = list()
count_vec_cosine_b = list()
lda_50topics_cosine_b = list()
lda_100topics_cosine_b = list()
lda_200topics_cosine_b = list()
labelb = list()

for x in range (1, len(df)):
    numbers = df['References_internal_clean'][x]
    numbers = list(map(int, numbers))
    for y in range (1, len(df)):
        if y in numbers:
            articles1.append(df['Article'][x])
            paragraphs1.append(df['Paragraphs_cleaned'][x])
            articles2.append(df['Article'][y])
            paragraphs2.append(df['Paragraphs_cleaned'][y])
            tfidf_cosine_a.append(float(cosine_similarity((npm_tfidf[x-1]), (npm_tfidf[y-1]))))
            tfidf_bigram_cosine_a.append(float(cosine_similarity((npm_tfidf_bigram[x-1]), (npm_tfidf_bigram[y-1]))))
            tfidf_trigram_cosine_a.append(float(cosine_similarity((npm_tfidf_trigram[x-1]), (npm_tfidf_trigram[y-1]))))
            count_vec_cosine_a.append(float(cosine_similarity((count_vectorizer[x-1]), (count_vectorizer[y-1]))))
            lda_50topics_cosine_a.append(float(cosine_similarity(matrix50[(x-1):x].values, matrix50[(y-1):y].values)))
            lda_100topics_cosine_a.append(float(cosine_similarity(matrix100[(x-1):x].values, matrix100[(y-1):y].values)))
            lda_200topics_cosine_a.append(float(cosine_similarity(matrix200[(x-1):x].values, matrix200[(y-1):y].values)))
            label.append(1)
        else:
            articles1b.append(df['Article'][x])
            paragraphs1b.append(df['Paragraphs_cleaned'][x])
            articles2b.append(df['Article'][y])
            paragraphs2b.append(df['Paragraphs_cleaned'][y])
            tfidf_cosine_b.append(float(cosine_similarity((npm_tfidf[x-1]), (npm_tfidf[y-1]))))
            tfidf_bigram_cosine_b.append(float(cosine_similarity((npm_tfidf_bigram[x-1]), (npm_tfidf_bigram[y-1]))))
            tfidf_trigram_cosine_b.append(float(cosine_similarity((npm_tfidf_trigram[x-1]), (npm_tfidf_trigram[y-1]))))
            count_vec_cosine_b.append(float(cosine_similarity((count_vectorizer[x-1]), (count_vectorizer[y-1]))))
            lda_50topics_cosine_b.append(float(cosine_similarity(matrix50[(x-1):x].values, matrix50[(y-1):y].values)))
            lda_100topics_cosine_b.append(float(cosine_similarity(matrix100[(x-1):x].values, matrix100[(y-1):y].values)))
            lda_200topics_cosine_b.append(float(cosine_similarity(matrix200[(x-1):x].values, matrix200[(y-1):y].values)))
            labelb.append(0)  
```
